```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
from IPython import display
from sklearn import datasets, preprocessing

(X, y) = datasets.make_circles(n_samples=1024, shuffle=True, noise=0.2, factor=0.4)
ind = np.logical_or(y==1, X[:,1] > X[:,0] - 0.5)
X = X[ind,:]
m = np.array([[1, 1], [-2, 1]])
X = preprocessing.scale(X)
y = y[ind]
y = 2*y - 1
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
h = 0.01
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
def visualize(X, y, w, loss, n_iter):
    plt.clf()
    Z = classify(np.c_[xx.ravel(), yy.ravel()], w)
    Z = Z.reshape(xx.shape)
    plt.subplot(1,2,1)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.subplot(1,2,2)
    plt.plot(loss)
    ymin, ymax = plt.ylim()
    plt.ylim(0, ymax)
    display.clear_output(wait=True)
    display.display(plt.gcf())
```

Implement classification with hinge loss.
You should use up to the second order terms of X for your classifier, as provided by function **expand**.

```python
def expand(X):
    X_ = np.zeros((X.shape[0], 6))
    X_[:,0:2] = X
    X_[:,2:4] = X**2
    X_[:,4] = X[:,0] * X[:,1]
    X_[:,5] = 1
    return X_

def classify(X, w):
    
    dotproduct = expand(X).dot(w)
    
    return np.sign(dotproduct)

def compute_loss(X, y, w):
    
    dotproduct = expand(X).dot(w)
    
    margin = dotproduct * y
    
    #warning, np.maximum is not same as np.max, read the description!
    hinge = np.maximum(0, 1 - margin)
    
    return np.mean(hinge)
    

def compute_grad(X, y, w):
    
    X_exp = expand(X)

    dotproduct = X_exp.dot(w)
    
    margin = dotproduct * y
    
    
    # if True, gradient is zero. (where hinge loss equals 0)
    is_zero =  1 - margin  <= 0
    
    #gradients for each sample of shape [n_samples, n_features]
    
    #margin = X*w * y , derivative of margin over W = X*y
    #derivative of max(1-margin,0) is -margin if 1-margin>0, else 0
    
    grad_term = -(X_exp*y[:,None]) * (~is_zero[:,None])
    #                     ^-- [:,None] transforms vector of [samples] into
    #                          a column-matrix [samples,1] in order for numpy
    #                          to multiply it over X_exp by broadcasting.
    
    #sum over samples batch
    return grad_term.sum(axis=0)
        
```

Find an optimal learning rate for gradient descent for given batch size.

```python
w = np.array([1,0,0,0,0,0])

alpha = 0.07 # learning rate

n_iter = 50
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))
for i in range(n_iter):
    ind = random.sample(range(X.shape[0]), batch_size)
    loss[i] = compute_loss(X, y, w)
    visualize(X[ind,:], y[ind], w, loss, n_iter)
    
    w = w - alpha * compute_grad(X[ind,:], y[ind], w)
    
    if i > 25:
        alpha *= 0.9

visualize(X, y, w, loss, n_iter)
plt.clf()
```

Implement RMSPROP algorithm


```python
w = np.array([0,0,0,0,0,1])
v = np.array([0,0,0,0,0,0])

alpha = 0.1 # learning rate. Please change.
mu    = 0.9 #moving average of gradient norm squared. Please change.
epsilon = 1e-8
ms = 0

n_iter = 50
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12,5))
for i in range(n_iter):
    ind = np.random.choice(X.shape[0], batch_size)
    loss[i] = compute_loss(X, y, w)
    visualize(X[ind,:], y[ind], w, loss, n_iter)
    
    grad = compute_grad(X, y, w)
    ms = mu*ms + (1-mu)*(grad*grad)
    coeff = alpha/(np.sqrt(ms + epsilon))
    w = w - coeff*grad


visualize(X, y, w, loss, n_iter)
plt.clf()
```

```python
# Adagrad
w = np.array([1,0,0,0,0,0])
v = np.array([0,0,0,0,0,0])

alpha = 0.2 # learning rate. Please change.
mu    = 0.9 # momentum
G = np.zeros_like(w, dtype=np.float64)

n_iter = 50
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))
for i in range(n_iter):
    ind = random.sample(range(X.shape[0]), batch_size)
    loss[i] = compute_loss(X, y, w)
    visualize(X[ind,:], y[ind], w, loss, n_iter)
    
    #update w here
    grad = compute_grad(X, y, w)
    G += grad**2
    w = w - (alpha/np.sqrt(G + 1e-8))*grad

visualize(X, y, w, loss, n_iter)
plt.clf()
```

```python
# Adadelta
w = np.array([1,0,0,0,0,0])
v = np.array([0,0,0,0,0,0])

alpha = 0.2 # learning rate. Please change.
mu    = 0.9 # momentum
G = np.zeros_like(w, dtype=np.float64)
G_delta = np.zeros_like(w, dtype=np.float64)
delta = 0

n_iter = 50
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))
for i in range(n_iter):
    ind = random.sample(range(X.shape[0]), batch_size)
    loss[i] = compute_loss(X, y, w)
    visualize(X[ind,:], y[ind], w, loss, n_iter)
    
    #update w here
    grad = compute_grad(X, y, w)
    G = mu*G + (1 - mu)*grad**2
    if i > 0:
        delta = (np.sqrt(G_delta + 1e-8)/np.sqrt(G + 1e-8))*grad
    else:
        delta = (alpha/np.sqrt(G + 1e-8))*grad
    G_delta = mu*G_delta + (1 - mu)*delta*delta 

    w = w - delta
    
    
visualize(X, y, w, loss, n_iter)
plt.clf()
```

```python
# Adam
w = np.array([1,0,0,0,0,0])
v = np.array([0,0,0,0,0,0])
mu = np.zeros_like(w, dtype=np.float64)

alpha = 0.1 # learning rate. Please change.
beta1 = 0.9
beta2 = 0.999
beta1_t = 1
beta2_t = 1


n_iter = 50
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))
for i in range(n_iter):
    ind = random.sample(range(X.shape[0]), batch_size)
    loss[i] = compute_loss(X, y, w)
    visualize(X[ind,:], y[ind], w, loss, n_iter)
    
    #update w here
    grad = compute_grad(X, y, w)
    mu = beta1*mu + (1 - beta1)*grad
    v = beta2*v + (1 - beta2)*grad*grad
    
    beta1_t *= beta1
    beta2_t *= beta2
    mu_c = mu / (1 - beta1_t)
    v_c = v / (1 - beta2_t)
    
    w = w - (alpha / (np.sqrt(v_c) + 1e-8)) * mu_c
    
    
visualize(X, y, w, loss, n_iter)
plt.clf()
```

```python
# Adamax
w = np.array([1,0,0,0,0,0])
v = np.array([0,0,0,0,0,0])
mu = np.zeros_like(w, dtype=np.float64)

alpha = 0.1 # learning rate. Please change.
beta1 = 0.9
beta2 = 0.999
beta1_t = 1
beta2_t = 1


n_iter = 50
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))
for i in range(n_iter):
    ind = random.sample(range(X.shape[0]), batch_size)
    loss[i] = compute_loss(X, y, w)
    visualize(X[ind,:], y[ind], w, loss, n_iter)
    
    #update w here
    grad = compute_grad(X, y, w)
    mu = beta1*mu + (1 - beta1)*grad
    v = np.maximum(beta2*v, np.abs(grad))
    
    beta1_t *= beta1
    beta2_t *= beta2
    mu_c = mu / (1 - beta1_t)
    
    w = w - (alpha / v) * mu_c
    
    
visualize(X, y, w, loss, n_iter)
plt.clf()
```
